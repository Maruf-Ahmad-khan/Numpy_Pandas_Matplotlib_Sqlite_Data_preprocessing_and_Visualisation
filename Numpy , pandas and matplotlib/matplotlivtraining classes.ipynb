```python
import matplotlib.pyplot as plt
import numpy as np
mu = 100
sigma = 15
x = mu + sigma * np.random.randn(1000)
n,bins,patches = plt.hist(x,20,facecolor = 'green',alpha =0.5)
#n is the highest peak point of each bar
#but the bins is use to show the middle point of x axis bar 
y = ((1/(np.sqrt(2 * np.pi * (sigma **2))))*np.exp(-0.5 * (1/sigma *(bins - mu ))**2))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
mu = 100
sigma = 15
x = mu + sigma * np.random.randn(1000)
n,bins,patches = plt.hist(x,20, density =1, facecolor = 'green',alpha =0.5)
#n is the highest peak point of each bar
#but the bins is use to show the middle point of x axis bar 
y = ((1/(np.sqrt(2 * np.pi * (sigma **2))))*np.exp(-0.5 * (1/sigma *(bins - mu ))**2))
plt.plot(bins,y,'b--')
plt.xlabel('x-axis')
plt.ylabel('Probability')
plt.title(r'Histogram of IQ: $\mu=100$,$\sigma =15$')
plt.show()
```
